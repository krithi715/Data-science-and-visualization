1. Rolling sensor analysis  

```python
print("Name: KRITHIKA K ")
print("Roll no : 24BAD061")

import numpy as np
import pandas as pd
sensor=np.array([22.1, 22.5, 23.0, 24.8, 25.1, 26.0, 27.5, 28.0])
s= pd.Series(sensor)
mean=s.rolling(3).mean()
std=s.rolling(3).std()
anomalies= s[abs(s-mean)>1.5*std]
print("Rolling Mean:\n",mean)
print("\nRolling Std:\n",std)
print("\nAnomalies:\n",anomalies)
```

2.Distance Matrix Computation

```python
print("Name: KRITHIKA K")
print("Roll no: 24BAD061")
import numpy as np
points=np.array([[2,3],[5,4],[8,7],[1,6]])
diff=points[:,np.newaxis,:]-points[np.newaxis,:,:]
distance_matrix= np.sqrt(np.sum(diff**2,axis=2))
print("Euclidean Distance Matrix:\n",distance_matrix)
```

3. Monte Carlo Risk Estimation

```python
print("Name: KRITHIKA K")
print("Roll no: 24BAD061")
import numpy as np
np.random.seed(10)
returns=np.random.normal(0.001,0.02,10000)
expected_return=np.mean(returns)
probablity_of_loss=np.mean(returns<0)
confidence_interval=np.percentile(returns,[2.5,97.5])
print("Expected Return:",expected_return)
print("Probablity of loss:",probablity_of_loss)
print("95% Confidence Interval:",confidence_interval)
```

4. Sparse Interaction Analysis

```python
print("Name: KRITHIKA K")
print("Roll no: 24BAD061")
import numpy as np
interactions= np.array([[0,0,5,0],[1,0,0,0],[0,0,0,3],[4,0,0,0]])
sparsity=(np.sum(interactions==0)/interactions.size)*100
row_sums=interactions.sum(axis=1,keepdims=True)
normalized=np.divide(interactions,row_sums,where=row_sums!=0)
top_interactions=interactions.max(axis=1)
print("Sparsity Percentage:",sparsity)
print("\nNormalised Matrix:\n",normalized)
print("\nTop Interaction Per Row:",top_interactions)
```

5.Student Performance Analysis

```python
print("Name: KRITHIKA K")
print("Roll no: 24BAD061")
import pandas as pd
data={
    "Student":["A","B","C","D","E"],
    "Math":[78,85,67,90,72],
    "Physics":[74,88,69,92,70],
    "Attendance":[82,90,76,95,80]
}

df=pd.DataFrame(data)
df["Average"]=df[["Math","Physics","Attendance"]].mean(axis=1)
df["Grade"]=df["Average"].apply(
    lambda x: "A" if x>=85 else("B" if x>=70 else "C")
)
top_performer=df.loc[df["Average"].idxmax()]
print("Student Performance Table:\n",df)
print("\nTop performer:\n",top_performer)
```

6.Time-Series Feature Engineering

```python
print("Name: KRITHIKA K")
print("Roll no: 24BAD061")
import pandas as pd
import numpy as np

timestamps=pd.date_range("2024-01-01",periods=10,freq="6h")
values=[12,18,20,22,19,15,14,17,21,23]
df=pd.DataFrame({"Timestamp":timestamps,"Value":values})

df["Hour"] = df["Timestamp"].dt.hour
df["Day"] = df["Timestamp"].dt.day

df["Hour_sin"]=np.sin(2* np.pi*df["Hour"]/24)
df["Hour_cos"]=np.cos(2*np.pi*df["Hour"]/24)
peak_hour=df.loc[df["Value"].idxmax(),"Hour"]

print("Time_series data:\n",df)
print("\nHour with peak activity:",peak_hour)
```

7. Data Leakage Detection

```python
print("Name: KRITHIKA K")
print("Roll no: 24BAD061")
import pandas as pd
train=pd.DataFrame({
    "id":[1,2,3,4],
    "score":[65,70,80,90],
    "result":[0,0,1,1]
})

test=pd.DataFrame({
    "id":[3,5],
    "score":[80,75],
    "result":[1,0]
})
overlap=pd.merge(train,test,on="id",how="inner")
leakage_columns=["result"] if "result" in test.columns else[]
leakage_percentage=(len(overlap)/len(test))*100
print("Overlapping Records:\n",overlap)
print("\nPotential Leakage colimns:",leakage_columns)
print("\nLeakage Percentage:",leakage_percentage,"%")
```

8. Memory Optimization

```python
print("Name: KRITHIKA K")
print("Roll no: 24BAD061")
import pandas as pd
import numpy as np
df=pd.DataFrame({
    "A":np.random.randint(0,1000,10000),
    "B":np.random.rand(10000),
    "C":["category"]*10000
})
memory_before=df.memory_usage(deep=True).sum()/1024
df["A"]=df["A"].astype("int16")
df["B"]=df["B"].astype("float32")
df["C"]=df["C"].astype("category")
memory_after=df.memory_usage(deep=True).sum()/1024

print("Memory usuage Before Optimization:{:.2f} KB".format(memory_before))
print("Memory usuage after optimization:{:.2f} KB".format(memory_after))
```

9.Uncertainty Visualization (Matplotlib)

```python
print("Name: KRITHIKA K")
print("Roll no: 24BAD061")
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(1,11)
y=np.array([10,12,15,18,20,22,24,23,25,27])
error=np.array([1,1.2,1.5,1.3,1.4,1.6,1.8,1.7,1.9,2.0])

plt.plot(x,y,label='Measurement')
plt.fill_between(x,y - error,y + error,alpha=0.2)
i=np.argmax(error)
plt.scatter(x[i],y[i],color='red')
plt.annotate(f'Max Uncertainty:{error[i]}',(x[i],y[i]),xytext=(x[i]+0.5,y[i]+1),
             arrowprops=dict(color='red'))
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line with Uncertainty')
plt.legend()
plt.show()
```

10. Correlation Heatmap (Matplotlib only)

```python
print("Name: KRITHIKA K")
print("Roll no: 24BAD061")
import pandas as pd
import matplotlib.pyplot as plt
df=pd.DataFrame({
    "Temp":[30,32,34,36,38],
    "Humidity":[70,65,60,55,50],
    "Pressure":[1012,1010,1008,1006,1004]
})
corr=df.corr()
plt.matshow(corr,cmap='coolwarm',vmin=-1,vmax=1)
plt.colorbar()
for i in range(len(corr)):
    for j in range(len(corr)):
        plt.text(j,i,f"{corr.iloc[i,j]:2f}",ha='center',va='center')
plt.xticks(range(len(corr.columns)),corr.columns)
plt.yticks(range(len(corr.columns)),corr.columns)
plt.show()
```
